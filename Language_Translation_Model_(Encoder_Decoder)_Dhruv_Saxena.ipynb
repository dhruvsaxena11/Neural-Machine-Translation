```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("rd124/marian-finetuned-kde4-en-to-hi")
model = AutoModelForSeq2SeqLM.from_pretrained("rd124/marian-finetuned-kde4-en-to-hi")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pl
from tensorflow import keras
```

```python
url = "https://raw.githubusercontent.com/suvaansh/Machine-Translation-English-to-Hindi-/master/hin.txt"
df = pd.read_table(url,header=None,names=["english","Hindi"])
df.head()
```

```python
df.shape
```

```python
#english tokens
from keras.preprocessing.text import Tokenizer
eng_tk=Tokenizer()
eng_tk.fit_on_texts(df.english)
len(eng_tk.word_counts)
```

```python
hin_tk=Tokenizer()
hin_tk.fit_on_texts(df.Hindi)
hin_tk.index_word[100]
```

```python
eng_x=eng_tk.texts_to_sequences(df.english)
eng_x
```

```python
hin_y=hin_tk.texts_to_sequences(df.Hindi)
len(hin_y)
```

```python
from keras.utils import pad_sequences
eng_x_new=pad_sequences(eng_x,padding="post")
hin_y_new=pad_sequences(hin_y,padding="post")
hin_y_new.shape
```

```python
my=hin_y_new.reshape(*hin_y_new.shape,1)
my[0]
```

```python
len(hin_y_new)
```

```python
len(hin_y_new[:, :-1][0])
```

```python
len(hin_y_new[:, 1:][0])
```

```python
model=keras.Sequential([
    keras.layers.Embedding(input_dim=2403,output_dim=400,input_length=len(eng_x_new[0])),
    keras.layers.LSTM(1000,return_sequences=False),
    keras.layers.RepeatVector(25),
    keras.layers.LSTM(1000,return_sequences=True,dropout=0.1),
    keras.layers.TimeDistributed(keras.layers.Dense(3012,activation="softmax"))


])

model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=["accuracy"])

history=model.fit(eng_x_new,hin_y_new,epochs=50)
```

```python
en=["name"]
en_new=eng_tk.texts_to_sequences(en)

en_x_new=pad_sequences(en_new,padding="post")
en_x_new
```

```python
# Assuming eng_testing[0] is a single English sequence
single_sequence = eng_testing[500]

# Reshape the sequence to have a batch size of 1
single_sequence = np.reshape(single_sequence, (1, len(single_sequence)))

# Get predictions
predictions = model.predict(np.array(single_sequence))
```

```python
predictions
```

```python
my=[]
for j in range(predictions.shape[1]):
    # Probabilities for the j-th token in the sequence
    probabilities_at_step = predictions[0, j, :]

    # Find the index of the token with the highest probability
    predicted_token_index = np.argmax(probabilities_at_step)

    # Print or use the predicted token index
    my.append(predicted_token_index)
```
